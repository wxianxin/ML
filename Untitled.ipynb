```python
import numpy as np
import timeit
from functools import partial
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()

def f_(x):
    x @ x
result = {}
for i in range(50, 70):
    h_size = i
    x = rng.standard_normal((h_size, h_size))
    func = partial(f_, x)
    result[h_size] = timeit.timeit(func, setup='', number=1000) * 1e6 / 1e3
```

```python
h_size = 64
x = rng.standard_normal((h_size, h_size))
%timeit f_(x)
```

```python
func = partial(f_, x)
execution_time = timeit.timeit(func, setup='', number=1000)
print("The code ran in", execution_time * 1e6, "us")
```

```python
fig = plt.figure(figsize=(8, 6), dpi=160)
ax = fig.add_subplot(111)
ax.plot(result.keys(), result.values(), color='lightblue', linewidth=3)
```
